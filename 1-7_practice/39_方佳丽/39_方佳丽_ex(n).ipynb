```python
from pony.orm import *


db = Database()


class Teacher(db.Entity):
    id = PrimaryKey(int, auto=True)
    name = Optional(str)
    passwd = Optional(str)
    no = Optional(str)
    courses = Set('Course')


class Course(db.Entity):
    id = PrimaryKey(int, auto=True)
    name = Optional(str)
    teacher = Required(Teacher)
    students = Set('Student')


class Student(db.Entity):
    id = PrimaryKey(int, auto=True)
    class = Required('Class')
    courses = Set(Course)
    scores = Set('Score')
    name = Optional(str)
    no = Optional(str)
    passwd = Optional(str)


class Class(db.Entity):
    id = PrimaryKey(int, auto=True)
    students = Set(Student)
    name = Optional(str)


class Quertion(db.Entity):
    id = PrimaryKey(int, auto=True)
    exams = Set('Exam')
    question = Optional(str)


class Exam(db.Entity):
    id = PrimaryKey(int, auto=True)
    scores = Set('Score')
    quertions = Set(Quertion)
    name = Optional(str)


class Score(db.Entity):
    id = PrimaryKey(int, auto=True)
    student = Required(Student)
    answers = Set('Answer')
    exam = Required(Exam)
    score = Optional(int)


class Answer(db.Entity):
    id = PrimaryKey(int, auto=True)
    score = Required(Score)
    question_id = Optional(str)
    answer = Optional(str)



db.generate_mapping()
```
